```python
# Import SQL Database
# Build connection through psycopg2
connection = psycopg2.connect("dbname=EmployeeSQL user=postgres password=******* host=localhost")

# Activate cursor connection
cur = connection.cursor()

# Execute table wanted from database
cur.execute('SELECT * FROM salaries')

# Fetch all into an assigned variable
salary_data = cur.fetchall()
salary_data
```

```python
# Put all the salary data in a list by looping it through the salary data
salary = []
for x in salary_data:
    salary.append(x[1])
    #print(x[1])
salary
```

```python
# Create a histogram to show the common salary ranges for the employees
plt.hist(salary)

plt.title("Histogram of Employee Salaries")
plt.xlabel("Salary")
plt.ylabel("Counts")

plt.show()
```

```python
# Importing a joined table from the database
cur.execute('SELECT t.title, ROUND(AVG(s.salary),2) as "Average Salary" FROM employees as e JOIN titles as t ON e.emp_title_id = t.title_id JOIN salaries as s ON e.emp_no = s.emp_no GROUP BY t.title;')
title_data = cur.fetchall()
title_data
```

```python
# Put all the titles in a list by looping through the title data
titles = []
for x in title_data:
    titles.append(x[0])
titles
```

```python
# Put all the average salaries in a list by looping through the title data
avg_salary = []
for x in title_data:
    avg_salary.append(float(x[1]))
avg_salary
```

```python
# Create a bar graph to show a comparison of average salary between the different titles
plt.bar(titles, avg_salary)
plt.xticks(rotation=30)

plt.title("Average Salary by Title")
plt.xlabel("Titles")
plt.ylabel("Salary Average")

plt.tight_layout()
plt.show()
```
